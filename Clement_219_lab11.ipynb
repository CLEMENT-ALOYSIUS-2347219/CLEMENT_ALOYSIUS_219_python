Q2. Perform the Exploratory Data Analysis on your domain-based dataset and demonstrate
the retrieved insights using “Matplotlib” modules. Visualize hidden insights using appropriate
plots (graphs) [Usage of line plot and scatter plot are mandatory]

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset (assuming it's in a CSV file)
df = pd.read_csv('monthly-milk-production-pounds.csv')

# Display the first few rows of your dataset to understand its structure
print(df.head())

# Summary statistics
print(df.describe())

# Scatter plot (Month vs. Monthly milk production)
plt.figure(figsize=(8, 6))
plt.scatter(df['Month'], df['Monthly milk production (pounds per cow)'], color='blue', label='Month vs. Monthly Production')
plt.title('Scatter Plot of Month vs. Monthly Milk Production')
plt.xlabel('Month')
plt.ylabel('Monthly Milk Production (pounds per cow)')
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()

# Line plot (Month vs. Monthly milk production)
plt.figure(figsize=(10, 6))
plt.plot(df['Month'], df['Monthly milk production (pounds per cow)'], marker='o', linestyle='-', color='green')
plt.title('Line Plot of Month vs. Monthly Milk Production')
plt.xlabel('Month')
plt.ylabel('Monthly Milk Production (pounds per cow)')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

# You can add more EDA and visualizations for other columns as needed
```
