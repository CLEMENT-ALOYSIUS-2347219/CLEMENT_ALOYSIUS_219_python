Q1. Apply regular expression for form validation. Create your domain-form using Tkinter
Module.
➔ Form should contain Text box [For Name, Email Id, Phone number], Dropdown [for
Gender], Spinbox [for Year/DoB] and other necessary widgets required for your
domain.
➔ Validate Your Name, Email Id, Phone number in the form.

```python
import tkinter as tk
from tkinter import ttk
import re

# Function to validate email using regular expression
def validate_email(email):
    # Regular expression for a valid email address
    pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
    if re.match(pattern, email):
        return True
    else:
        return False

# Function to validate phone number using regular expression
def validate_phone(phone):
    # Regular expression for a valid phone number (10 digits)
    pattern = r'^\d{10}$'
    if re.match(pattern, phone):
        return True
    else:
        return False

# Function to submit the form
def submit_form():
    name = name_entry.get()
    email = email_entry.get()
    phone = phone_entry.get()
    gender = gender_var.get()
    dob = dob_spinbox.get()

    # Validation
    if not name:
        result_label.config(text="Name is required", fg="red")
    elif not validate_email(email):
        result_label.config(text="Invalid Email", fg="red")
    elif not validate_phone(phone):
        result_label.config(text="Invalid Phone Number", fg="red")
    else:
        # Form data is valid, you can process it here
        result_label.config(text="Form submitted successfully", fg="green")
        # Example: Print the form data
        print("Name:", name)
        print("Email:", email)
        print("Phone:", phone)
        print("Gender:", gender)
        print("DOB:", dob)

# Create the main window
root = tk.Tk()
root.title("Milk Dairy Management Form")

# Name Label and Entry
name_label = ttk.Label(root, text="Name:")
name_label.grid(row=0, column=0, padx=10, pady=10, sticky="w")
name_entry = ttk.Entry(root)
name_entry.grid(row=0, column=1, padx=10, pady=10)

# Email Label and Entry
email_label = ttk.Label(root, text="Email:")
email_label.grid(row=1, column=0, padx=10, pady=10, sticky="w")
email_entry = ttk.Entry(root)
email_entry.grid(row=1, column=1, padx=10, pady=10)

# Phone Label and Entry
phone_label = ttk.Label(root, text="Phone:")
phone_label.grid(row=2, column=0, padx=10, pady=10, sticky="w")
phone_entry = ttk.Entry(root)
phone_entry.grid(row=2, column=1, padx=10, pady=10)

# Gender Label and Dropdown
gender_label = ttk.Label(root, text="Gender:")
gender_label.grid(row=3, column=0, padx=10, pady=10, sticky="w")
gender_var = tk.StringVar()
gender_dropdown = ttk.Combobox(root, textvariable=gender_var, values=["Male", "Female", "Other"])
gender_dropdown.grid(row=3, column=1, padx=10, pady=10)

# Date of Birth Label and Spinbox
dob_label = ttk.Label(root, text="Date of Birth (Year):")
dob_label.grid(row=4, column=0, padx=10, pady=10, sticky="w")
dob_spinbox = ttk.Spinbox(root, from_=1900, to=2023)
dob_spinbox.grid(row=4, column=1, padx=10, pady=10)

# Submit Button
submit_button = ttk.Button(root, text="Submit", command=submit_form)
submit_button.grid(row=5, columnspan=2, padx=10, pady=20)

# Result Label
result_label = ttk.Label(root, text="", font=("Helvetica", 12))
result_label.grid(row=6, columnspan=2)

root.mainloop()
```
